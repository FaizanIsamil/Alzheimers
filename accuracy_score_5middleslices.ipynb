```python
from statistics import mode
```

```python
def most_common(List):
    return(mode(List))
```

```python
k=0
pred_labels=[]
for i in range(469):
    j=k+5
    md=most_common(dff[k:j])
    pred_labels.append(md)
    k=k+5
```
